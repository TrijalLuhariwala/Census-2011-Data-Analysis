# Column Reduction for easier Data Handling

```python
colsdrop = ['srcPC11DistrictCode', 'srcPC11SubDistrictCode', 'srcPC11TownVillageCode', 'srcPC11TownVillageName','srcYear', 'YearCode',
                'srcPC11StateCode','Year']
df = df.drop(columns=colsdrop)
```

### Merge Columns

```python
df['Gender_Ratio'] = (df['Female population'] / df['Male population']) * 1000
df['Child_Population_Ratio'] = (df['Population in the age group 0 to 6 years'] / df['Population']) * 100
df['SC_ST_Population'] = df['Scheduled Caste population'] + df['Scheduled Tribe population']
df['Literacy_Rate'] = (df['Literate population '] / (df['Literate population '] + df['Illiterate population'])) * 100
df['Female_Literacy_Rate'] = (df['Female literate population '] / (df['Female population'])) * 100
```

```python
colsdrop2= ['Female population','Male population','Population in the age group 0 to 6 years','Scheduled Caste population','Scheduled Tribe population',
            'Literate population ','Illiterate population','Female literate population ']
```

```python
df = df.drop(columns=colsdrop2)
```

```python
df['Main_Workers_Agriculture']=df['Number of main workers as cultivators']+df['Number of main workers as agricultural labourers']
df['Marginal_Workers_Agriculture']=df['Number of marginal workers as cultivators']+df['Number of marginal workers as agricultural labourers']
```

```python
colsdrop3=['Number of main workers as cultivators','Number of main workers as agricultural labourers','Number of marginal workers as cultivators',
          'Number of marginal workers as agricultural labourers']
df=df.drop(columns=colsdrop3)
```

```python
final=['TRU', 'Households','Population','Gender_Ratio', 'Child_Population_Ratio','SC_ST_Population','Literacy_Rate', 'Female_Literacy_Rate',
    'Working population', 'Main_Workers_Agriculture','Marginal_Workers_Agriculture','Number of main workers in the household industry',
    'Number of  main workers as Other workers','Number of marginal workers in the household industry','Number of marginal workers as Other workers',
    'Number of Non workers']
df = df[final]
```

# Outliers Treatment

```python
df.boxplot()
```

```python
cols=df.select_dtypes(include=[np.number]).columns.tolist()
for col in cols:
    df[col] = df[col].astype(float)
    Q1 = np.percentile(df[col], 25)
    Q3 = np.percentile(df[col], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    df[col] = np.where(df[col] < lower_bound, lower_bound,df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound,df[col])
```
